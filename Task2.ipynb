```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
```

```python
df1 = pd.read_csv("Unemployment in India.csv")
df2 = pd.read_csv("Unemployment_Rate_upto_11_2020.csv")
```

```python
print("Dataset 1")
print(df1.head())
print("\nDataset 2")
print(df2.head())
```

```python
print("Dataset 1 Info")
print(df1.info())
print("\nDataset 2 Info")
print(df2.info())
```

```python
print("Dataset 1 Null Values")
print(df1.isnull().sum())
print("\nDataset 2 Null Values")
print(df2.isnull().sum())
```

```python
df1.columns = [
    'Region',
    'Date',
    'Frequency',
    'Estimated_Unemployment_Rate',
    'Estimated_Employed',
    'Estimated_Labour_Participation_Rate',
    'Area'
]
df2.columns = [
    'Region',
    'Date',
    'Frequency',
    'Estimated_Unemployment_Rate',
    'Estimated_Employed',
    'Estimated_Labour_Participation_Rate',
    'Area',
    'Longitude',
    'Latitude'
]
print(df1.columns)
print(df2.columns)
```

```python
# Convert date column
df1['Date'] = pd.to_datetime(df1['Date'])
df2['Date'] = pd.to_datetime(df2['Date'])
print(df1['Date'].head())
```

```python
print(df1.describe())
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df1['Estimated_Unemployment_Rate'], bins=20, kde=True)
plt.title("Distribution of Unemployment Rate")
plt.xlabel("Unemployment Rate")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.barplot(
    x='Region',
    y='Estimated_Unemployment_Rate',
    data=df1
)
plt.xticks(rotation=90)
plt.title("Average Unemployment Rate by Region")
plt.xlabel("Region")
plt.ylabel("Unemployment Rate")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(
    x='Area',
    y='Estimated_Unemployment_Rate',
    data=df1
)
plt.title("Urban vs Rural Unemployment Rate")
plt.show()
```

```python
# Monthly average unemployment
monthly_data = df1.groupby(df1['Date'].dt.month)[
    'Estimated_Unemployment_Rate'
].mean()
plt.figure(figsize=(10,5))
monthly_data.plot(marker='o')
plt.title("Monthly Unemployment Trend")
plt.xlabel("Month")
plt.ylabel("Average Unemployment Rate")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.lineplot(
    x='Date',
    y='Estimated_Unemployment_Rate',
    data=df1
)
plt.title("Unemployment Rate During Covid-19")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate")
plt.show()
```

```python
top_regions = df1.groupby('Region')[
    'Estimated_Unemployment_Rate'
].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,5))
top_regions.plot(kind='bar')
plt.title("Top 10 Regions with Highest Unemployment")
plt.xlabel("Region")
plt.ylabel("Average Unemployment Rate")
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.barplot(
    x='Region',
    y='Estimated_Employed',
    data=df1
)
plt.xticks(rotation=90)
plt.title("Estimated Employment by Region")
plt.show()
```

```python
area_counts = df1['Area'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(
    area_counts,
    labels=area_counts.index,
    autopct='%1.1f%%'
)
plt.title("Urban vs Rural Distribution")
plt.show()
```
